```python
import pandas as pd
import os

# Caminho base
caminho = 'E:\\Projetos\\ABMS-WP' # Ajuste se necessário

# 1. Carregar arquivo original
path_orig = os.path.join(caminho, 'includes', 'Tabela_consumidores_Itapua_com_setor.csv')
df_orig = pd.read_csv(path_orig, sep=',')
print(f"1. Total de consumidores no arquivo ORIGINAL: {len(df_orig)}")

# 2. Verificar quantos têm Setor preenchido
df_com_setor = df_orig.dropna(subset=['CD_SETOR'])
df_com_setor = df_com_setor[df_com_setor['CD_SETOR'].astype(str) != '']
print(f"2. Total de consumidores COM SETOR VÁLIDO: {len(df_com_setor)}")

# 3. Verificar arquivo final gerado
path_final = os.path.join(caminho, 'includes', 'Tabela_consumidores_Itapua_com_setor_comportamento_e_renda.csv')
if os.path.exists(path_final):
    df_final = pd.read_csv(path_final, sep=';')
    print(f"3. Total de linhas no arquivo FINAL (que vai pro GAMA): {len(df_final)}")
    
    # Verificar soma do consumo
    if 'NN_MEDIA_CONSUMO' in df_final.columns:
        print(f"4. Soma total do consumo (m³) no arquivo final: {df_final['NN_MEDIA_CONSUMO'].sum():.2f}")
else:
    print("Arquivo final não encontrado.")
```
