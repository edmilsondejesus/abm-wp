```python
import pandas as pd
import os

# Define paths
# path = 'C:\\Users\\e124796\\Gama_Workspace\\ABMS-WP'
# path = 'C:\\Users\\Edmilson\\Gama_Workspace2\\ABMS-WP'
path = 'E:\\Projetos\\ABMS-WP' # Last one active

# 1. Load the .csv file with CD_SETOR and SK_MATRICULA columns
df = pd.read_csv(os.path.join(path, 'includes', 'Tabela_consumidores_Itapua_com_setor.csv'), sep=',')

# Extract the last 3 characters from the CD_SETOR column
# Adds 'S' prefix to match GAMA format
df['CD_SETOR'] = 'S' + df['CD_SETOR'].astype(str).str[7] + df['CD_SETOR'].astype(str).str[-2:]

# 2. Quantify the number of connections (SK_MATRICULA) per sector (CD_SETOR)
connections_by_sector = df.groupby('CD_SETOR')['SK_MATRICULA'].count().reset_index()
connections_by_sector.rename(columns={'SK_MATRICULA': 'NUM_CONNECTIONS'}, inplace=True)

# 3. Sort sectors by number of connections (from highest to lowest)
connections_by_sector = connections_by_sector.sort_values(by='NUM_CONNECTIONS', ascending=False)

# 4. Select the top 10 sectors
top_10_sectors = connections_by_sector.head(10)

# 5. Save the result to a new CSV file (optional)
top_10_sectors.to_csv(os.path.join(path, 'resultados', 'Tabela_top_10_setores.csv'), index=False)

# 6. View the result
print(top_10_sectors)
```
