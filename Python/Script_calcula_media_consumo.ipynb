```python
import pandas as pd
import os

# Define the base project directory
# Uncomment the appropriate path for your environment
base_path = 'E:\\Projetos\\ABM-WP' # Currently active path

# Define input and output file paths
input_csv = os.path.join(base_path, 'includes', 'Tabela_consumo_Itapua_60m.csv')
output_csv = os.path.join(base_path, 'includes', 'Tabela_consumo_medio_Itapua_12m.csv')

# Load the CSV file
# delimiter=';' assumes the CSV uses semi-colons as separators
df = pd.read_csv(input_csv, delimiter=';')

# Sort the dataframe
# Primary sort: 'SK_MATRICULA' (User ID) -> Ascending
# Secondary sort: 'AM_REFERENCIA' (Reference Month/Year) -> Descending (to get the most recent dates first)
df = df.sort_values(by=['SK_MATRICULA', 'AM_REFERENCIA'], ascending=[True, False])

# Select the most recent 12 records (months) for each unique ID
df = df.groupby('SK_MATRICULA').head(12)

# Calculate the mean of 'HCLQTCON' (Consumption Volume) for each 'SK_MATRICULA'
mean_consumption_by_id = df.groupby('SK_MATRICULA')['HCLQTCON'].mean().reset_index()

# Save the processed data to a new CSV file
mean_consumption_by_id.to_csv(output_csv, index=False, sep=';')

print(f"File successfully saved to: {output_csv}")
```
