```python
import pandas as pd
import os

# ==========================================
# 1. Configuration and Paths
# ==========================================
BASE_PATH = r'E:\Projetos\ABM-WP'
INPUT_FILE = os.path.join(BASE_PATH, 'resultados','consumo_previsto_todos_cenarios.csv')
OUTPUT_FILE = os.path.join(BASE_PATH, 'resultados','matriz_consumo_anual_por_cenario.csv')

def process_annual_aggregation():
    """
    Aggregates monthly values into annual totals, rounds to 1 decimal place,
    and adds the m3 unit to the scenario index.
    """
    print(f"--- Processing: {INPUT_FILE} ---")
    
    if not os.path.exists(INPUT_FILE):
        print(f"Error: {INPUT_FILE} not found.")
        return

    # Load data with decimal comma handling
    df = pd.read_csv(INPUT_FILE, sep=';', decimal=',', low_memory=False)

    # ==========================================
    # 2. Aggregation and Rounding
    # ==========================================
    # Group by Year and sum scenarios
    annual_data = df.drop(columns=['Mes', 'Mes_Ano']).groupby('Ano').sum()

    # Transpose to have Scenarios as rows and Years as columns
    final_matrix = annual_data.T
    
    # NEW: Rounding values to 1 decimal place for academic precision
    final_matrix = final_matrix.round(1)

    # NEW: Renaming the index to clearly state the unit (m3)
    final_matrix.index.name = 'Scenario (m³)'

    # ==========================================
    # 3. Export
    # ==========================================
    # Exporting with comma as decimal for consistency or dot for international journals
    # For Elsevier, dot (.) is standard for decimals:
    final_matrix.to_csv(OUTPUT_FILE, decimal='.', sep=',')
    
    print(f"--- Success! Matrix saved at: {OUTPUT_FILE} ---")
    print("\nPreview of Annual Results (rounded to 1 decimal):")
    print(final_matrix.head())

if __name__ == "__main__":
    process_annual_aggregation()
```
